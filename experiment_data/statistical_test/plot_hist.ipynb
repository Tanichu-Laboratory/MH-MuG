# Plot Histgram

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
#import xlsx file
result_dir = "result_masterthese"
xlsx_paths = sorted(glob.glob(f"{result_dir}/*.xlsx"))
dfs = []
for path in xlsx_paths:
    dfs.append(pd.read_excel(path, header=0, index_col=0))
```

```python
dfs[0].index
```

```python
#plot
plt.rcParams["font.size"] = 18
groups = dfs[0].index
x = range(0, 5)
x_label = ["1", "2", "3", "4", "5"]
y_min = 0
y_max = 40
fig, axs = plt.subplots(len(groups), len(dfs), figsize=(12, 12), sharex=True, sharey=True)
ax = [[]for i in range(len(groups))]
color = ["red", "blue", "green"]
for group_idx in range(len(groups)):
    #ser graph each group
    for df_idx, df in enumerate(dfs):
        ax[group_idx].append(axs[group_idx, df_idx])
        ax[group_idx][df_idx].bar(x, df.iloc[group_idx], tick_label=x_label, color=color[df_idx])
        plt.ylim(y_min, y_max)

plt.subplots_adjust(wspace=0, hspace=0)
plt.tight_layout()
plt.show()
```
